```python
import requests
from bs4 import BeautifulSoup

def get_etf_holdings(etf_code):
    url = f"https://www.moneydj.com/ETF/X/Basic/ETF/{etf_code}/Holdings"
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    table = soup.find('table', {'class': 'datalist'})
    rows = table.find_all('tr')
    
    holdings = []
    for row in rows[1:]:
        cols = row.find_all('td')
        stock_code = cols[0].text.strip()
        stock_name = cols[1].text.strip()
        percentage = cols[2].text.strip()
        
        holding = {
            'stock_code': stock_code,
            'stock_name': stock_name,
            'percentage': percentage
        }
        holdings.append(holding)
    
    return holdings

# 範例用法
etf_code = '0050'  # 以元大台灣50為例
holdings = get_etf_holdings(etf_code)

print(f"ETF代號: {etf_code}")
for holding in holdings:
    print(f"股票代號: {holding['stock_code']}, 股票名稱: {holding['stock_name']}, 持股比例: {holding['percentage']}")
```
